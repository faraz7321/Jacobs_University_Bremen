```python
import numpy as np
import pandas as pd
import seaborn as sb
from sklearn import tree
from sklearn import ensemble
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
dataset = pd.read_csv('data.csv')
y = dataset['Diagnosis'] # Dedendent Variable
x = dataset.drop(['ID ','Diagnosis'], axis = 1) # Independent Variable
dataset.head()
```

```python
x.isnull().sum()
```

```python
# independent = dataset.iloc[:,2:29]
# dependent = dataset.iloc[:,1]
x.head()
```

```python
# X and Y variables for test and train

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=2) # Test Size is 25%
```

```python
# x_train
# x_test
# y_train
# y_test
```

```python
# Simple Decision Tree

DTclf = DecisionTreeClassifier(max_depth=5)
DTclf.fit(x_train,y_train) # Train
DTclf.score(x_test,y_test) # Test
y_pred = DTclf.predict(x_test)
DTclf.score(x_test,y_test)
```

```python
# Random Forest Classifier

RFclf = ensemble.RandomForestClassifier(n_estimators=150)
RFclf.fit(x_train,y_train) # Train
RFclf.score(x_test,y_test) # Test
```

```python
# Logistic Regression

LRmodel = LogisticRegression(max_iter=500)
# clf.fit(x_train,y_train)
LRmodel.fit(x_train,y_train) # Train
LRmodel.score(x_test,y_test) # Test
```
