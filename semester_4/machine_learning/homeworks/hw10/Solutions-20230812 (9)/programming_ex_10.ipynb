```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist

(X_train,G_train),(X_test,G_test) = mnist.load_data()
idx_4 = np.where(G_train==8)
X=X_train[idx_4,:,:]
X = X[0,:,:,:]

X = X.reshape(X.shape[0],28*28)

fig, axs = plt.subplots(1, 4,sharex=True, sharey=True)
for i in range(4):
    axs[i].imshow(X[i,:].reshape((28,28)), cmap="gray")
    
```

For image compression (reducing the dimension of an image from a higher value D to a lower value d), we will find the d principal components and project the image onto these dimensions. To decompress or view the image again, we can multiply the compressed image with V matrix again (keeping dimensions and multiplying order in mind).

Principal componends can be obtained by PCA or SVD. First d columns of V matrix from SVD are the principal components.

```python
U,D,V_transp = np.linalg.svd(X,full_matrices=False)
V=V_transp.T
for d in [784, 512, 256, 128, 64, 32]:
    print("d=%d"%(d))
    V_d = V[:,:d]
    x=[]
    x_tilde=[]
    fig, axs = plt.subplots(1, 4,sharex=True, sharey=True)
    for i in range(4):
        compressed_img = V_d.T @ X[i,:]
        decompressed_img = V_d @ compressed_img
        axs[i].imshow(decompressed_img.reshape((28,28)), cmap="gray")
        axs[i].set_title("d=%d"%(d))
```
